**Dropping unwamted columns**

```python
data = data.drop(columns="State Code")
data
```

**Transforming data types**

```python
data.dtypes
```

```python
data["County Name"] = data["County Name"].astype("string")
data["State"] = data["State"].astype("string")
data["SVI Category"] = data["SVI Category"].astype("category")
data["CVAC Level Of Concern"] = data["CVAC Level Of Concern"].astype("category")
data.dtypes
```

**Dealing with NULL values**

```python
data.isna().sum()
```

```python
data["Percent adults fully vaccinated against COVID-19 (as of 6/10/21)"].fillna(0)
```

**Renaming columns**

```python
data = data.rename(columns={"Percent non-Hispanic American Indian/Alaska Native": "Percent non-Hispanic (American Indian/Alaska Native)"})
data = data.rename(columns={"Percent non-Hispanic Black": "Percent non-Hispanic (Black)"})
data = data.rename(columns={"Percent non-Hispanic Asian": "Percent non-Hispanic (Asian)"})
data = data.rename(columns={"Percent non-Hispanic Native Hawaiian/Pacific Islander": "Percent non-Hispanic (Native Hawaiian/Pacific Islander)"})
data = data.rename(columns={"Percent non-Hispanic White": "Percent non-Hispanic (White)"})
data
```

**Creating a calculated field based on two other columns**

```python
# Calculate Average Hesitancy Level

data['Average Hesitancy Level'] = data[['Estimated hesitant', 'Estimated hesitant or unsure', 'Estimated strongly hesitant']].mean(axis=1)
average_hesitancy_level = data.pop('Average Hesitancy Level')
data.insert(6, 'Average Hesitancy Level', average_hesitancy_level)

data
```

**Removing the duplicate rows**

```python
data.drop_duplicates()
```

**Handling Outliers**

```python
import seaborn as sns

sns.boxplot(data["Average Hesitancy Level"])
```

```python
import numpy as np

Q1 = np.percentile(data["Average Hesitancy Level"], 25)
Q3 = np.percentile(data["Average Hesitancy Level"], 75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

data = data[data["Average Hesitancy Level"] < upper_bound]
data = data[data["Average Hesitancy Level"] > lower_bound]


sns.boxplot(data["Average Hesitancy Level"])
```
